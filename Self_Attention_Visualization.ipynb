```python
!pip install -q transformers
!pip install -q bertviz

from transformers import BertModel, BertTokenizer
from bertviz import head_view
import torch

model = BertModel.from_pretrained("bert-base-uncased", output_attentions=True)
tokenizer = BertTokenizer.from_pretrained("bert-base-uncased")

model.eval()
sentence = "The bank was next to the river"

inputs = tokenizer.encode_plus(sentence, return_tensors="pt", add_special_tokens=True)
input_ids = inputs["input_ids"]
token_type_ids = inputs["token_type_ids"]
with torch.no_grad():
    outputs = model(input_ids=input_ids, token_type_ids=token_type_ids)
    attention = outputs.attentions

tokens = tokenizer.convert_ids_to_tokens(input_ids[0])
head_view(attention, tokens)
```

```python
layer_num = 0
head_num = 0
attn_matrix = attention[layer_num][0, head_num]
tokens = tokenizer.convert_ids_to_tokens(input_ids[0])

# Print attention scores between words
print(f"Attention scores from Layer {layer_num}, Head {head_num}:\n")
for i, token_from in enumerate(tokens):
    for j, token_to in enumerate(tokens):
        score = attn_matrix[i][j].item()
        print(f"{token_from:>10}  {token_to:>10} : {score:.4f}")
    print()
```
