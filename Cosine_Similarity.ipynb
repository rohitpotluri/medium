```python
!pip install -q gensim

from gensim.downloader import load
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

glove_vectors = load('glove-wiki-gigaword-100')

apple_vec = glove_vectors['apple']
fruit_vec = glove_vectors['fruit']
television_vec = glove_vectors['television']

sim_apple_fruit = cosine_similarity([apple_vec], [fruit_vec])[0][0]
sim_apple_television = cosine_similarity([apple_vec], [television_vec])[0][0]

print("Cosine similarity (apple, fruit):", sim_apple_fruit)
print("Cosine similarity (apple, television):", sim_apple_television)
```
